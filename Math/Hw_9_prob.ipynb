```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```



Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.


```python
fig = plt.figure(figsize=(15, 10), dpi=80)
#t = plt.title('Distributions of sample means from exponential distribution')

rv = sts.expon()
ns = [5, 10, 50, 100, 500, 1000,]
for i, n in enumerate(ns):
    sample = rv.rvs(size=1000*n)
    sample.shape = n, 1000
    means = np.mean(sample, axis=0)
    
    ax = fig.add_subplot(231+i)
    x_1 = ax.hist(
        means,
        bins=50,
        #weights = np.ones_like(means) / (len(means)),
        # range=(0, 3),
        # density=False,
        # stacked=True,
        label='1000 samples of\n'+str(n)+ ' elements'
    )
    
    ax.set_xlabel('Sample mean')
    ax.set_ylabel('Frequency')
    ax.legend()

    
```
