#### Задание 1¶

Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, t
```

```python
fig = plt.figure(figsize=(15, 5), dpi=80)
ax = fig.add_subplot(111)
sigmas = [0.5, 1, 2, 5, 50]
colors = ('red', 'orange', 'green', 'brown', 'black')

x = np.linspace( -10, 10, 200
    #t.ppf(0.01, df),
    #t.ppf(0.99, df), 100
)

for s, color in zip(sigmas, colors):
    ax.plot(
        x, norm.pdf(x, scale=s),
        'r-', lw=2, alpha=0.6,
        color=color,
        label='scale parameter = '+str(s)
    )
l = ax.legend(title='Degrees of freedom')
tl = plt.title('normal distributions with different scale parameters')
xl = ax.set_xlabel('x')
yl = ax.set_ylabel('probability dencity')
```

```python
fig = plt.figure(figsize=(15, 5), dpi=80)
ax = fig.add_subplot(111)
lambdas = [0.5, 1, 2, 5, 7]
colors = ('red', 'orange', 'green', 'brown', 'black')

x = np.linspace( -0.1, 5, 200
    #t.ppf(0.01, df),
    #t.ppf(0.99, df), 100
)

for lambda_, color in zip(lambdas, colors):
    ax.plot(
        x, expon.pdf(x, scale=1/lambda_),
        'r-', lw=2, alpha=0.6,
        color=color,
        label='lambda = '+str(lambda_)
    )
l = ax.legend(title='Degrees of freedom')
tl = plt.title('exponential distributions with different scale parameters')
xl = ax.set_xlabel('x')
yl = ax.set_ylabel('probability dencity')
```

```python
fig = plt.figure(figsize=(15, 5), dpi=80)
ax = fig.add_subplot(111)
dfs = [1, 2, 3, 5, 50]
colors = ('red', 'orange', 'green', 'brown', 'black')

x = np.linspace( -10, 10, 200
    #t.ppf(0.01, df),
    #t.ppf(0.99, df), 100
)

for df, color in zip(dfs, colors):
    ax.plot(
        x, t.pdf(x, df),
        'r-', lw=2, alpha=0.6,
        color=color,
        label='mu = '+str(df)
    )
l = ax.legend(title='Degrees of freedom')
tl = plt.title('t-distributions with different degrees of freedom')
xl = ax.set_xlabel('x')
yl = ax.set_ylabel('probability dencity')
```

Непрерывная случайная величина задана плотностью распределения: $$ f(\xi) = C, \xi \in [0;5];f(\xi) = 0, \xi \notin [0;5]$$ Найдите C, математическое ожидание $\xi$ а также вероятность попадания $\xi$ в отрезок $[3.5;7]$, т.е.
$ P(3.5 \lt \xi \lt 7) $.


$ \int_{-\inf}^{+\inf}C d\xi = \int_{0}^{5}Cd\xi = 5 \cdot C = 1$

$ C = 1/5 = 0.2 $

$ E[\xi] = 0.25 $

$ P(\,3.5 \lt \xi \lt 7\,) = (5-3.5)/5 = 1.5/5 = 0.3 $

#### Задание 3

Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?


светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным

```python
full_t = 54 + 4 + 28
p_red = 54/full_t; p_amber = 4/full_t; p_green = 28/full_t
from math import log
info = -p_red*log(p_red, 2) + -p_amber*log(p_amber, 2) + -p_green*log(p_green, 2)
print(round(info, 2), 'bit')
```

 в случае 28, 28 и 28 секунд для каждого цвета

```python
p2 = 1/3
info = -3*p2*log(p2, 2)
print(round(info, 2), 'bit')
```
