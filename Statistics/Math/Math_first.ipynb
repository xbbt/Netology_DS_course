### Уровень 0:
Даны вектора x=(1,1), y=(2, 0) и z = (0,2)
Надо:
Изобразить вектора на экране
Изобразить точку x + y + z
Найти угол между векторами x и y

```python
x = (1, 1)
y = (2, 0)
z = (0, 2)
a = x[0] + y[0] + z[0], x[1] + y[1] + z[1]
```

```python
import matplotlib.pyplot as plt
ax = plt.axes()
ax.arrow(0, 0, *x, head_width=0.1, head_length=0.1, fc='lightblue', ec='black')
ax.arrow(0, 0, *y, head_width=0.1, head_length=0.1, fc='lightblue', ec='black')
ax.arrow(0, 0, *z, head_width=0.1, head_length=0.1, fc='lightblue', ec='black')
ax.arrow(0, 0, *z, head_width=0.1, head_length=0.1, fc='lightblue', ec='black')
ax.arrow(0, 0, *a, head_width=0.1, head_length=0.1, fc='lightblue', ec='black')

plt.xlim(-0.5, 3.5)
plt.ylim(-0.5, 3.5)
plt.grid()
```

```python
from math import acos, degrees
degrees(acos((x[0]*y[0] + x[1]*y[1])/(x[0]**2 + x[1]**2)**0.5/(y[0]**2+y[1]**2)**0.5))
```
