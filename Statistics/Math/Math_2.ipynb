#### Уровень 1:
Дан объект в 2d пространстве:

```python
import numpy as np
import matplotlib.pyplot as plt

A = [
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
]
A = np.array(A)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

При помощи матричных операций выполните:

 -уменьшить объект в два раза.

```python
change = 0.5*np.eye(6)
A2 = np.dot(change, A)
x = A2[:,0]
y = A2[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Повернуть на 130 градусов.

```python
from math import radians, cos, sin
def rotation_matrix(alpha=130):
    alpha = radians(alpha)
    return  np.matrix([[cos(alpha), -sin(alpha), 0], [sin(alpha), cos(alpha), 0], [0, 0, 1]])
A2 = np.dot(A, rotation_matrix())
x = A2[:,0]
y = A2[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Отразить объект относительно прямой y=x

```python
mirror = np.matrix([[0, 1, 0], [1, 0, 0], [0, 0, 1]])
A2 = np.dot(A, mirror)
x = A2[:,0]
y = A2[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций).

смещение на (200, 300) и
масштабирование на (0.5, 1,2)

```python
move = np.matrix([[1, 0, 0], [0, 1, 0], [200, 300, 1]])
scale = np.matrix([[0.5, 0, 0], [0, 1.2, 0], [0, 0, 1]])
A2 = np.dot(np.dot(A, move), scale)
x = A2[:,0]
y = A2[:,1]
plt.plot(x, y)
plt.ylim([-500, 500])
plt.xlim([-500, 500])
```
